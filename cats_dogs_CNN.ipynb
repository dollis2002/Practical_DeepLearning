```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import pickle
import numpy as np

# Import data from pickled sets in cats_dogs.ipynb
X = pickle.load(open('X.pickle', 'rb'))
y = pickle.load(open('y.pickle', 'rb'))

# Normalize
X=np.array(X/255.0)
y=np.array(y)

model = Sequential([
    Conv2D(64, (3,3), input_shape=X.shape[1:]),
    Activation("relu"),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(64, (3,3)),
    Activation("relu"),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(64),
    Activation("relu"),
    Dense(1),
    Activation('sigmoid')
])

model.compile(
    loss='binary_crossentropy', 
    optimizer='adam', 
    metrics=['accuracy']
)

model.fit(X, y, batch_size=32, validation_split=0.1, epochs=10)
```

```python
# Save model
model.save('cats_dogs.model')
```
