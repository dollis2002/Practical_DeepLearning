```python

import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import TensorBoard
import pickle
import numpy as np
import time

gpu_options = tf.compat.v1.GPUOptions(per_process_gpu_memory_fraction=0.3)
sess = tf.compat.v1.Session(config=tf.compat.v1.ConfigProto(gpu_options=gpu_options))

# Import data from pickled sets in cats_dogs.ipynb
X = pickle.load(open('X.pickle', 'rb'))
y = pickle.load(open('y.pickle', 'rb'))

# Normalize
X=np.array(X/255.0)
y=np.array(y)

# Test model with diferent layer combinations
dense_layers = [1]
layer_sizes = [64]
conv_layers = [3]

for dense in dense_layers:
    for layer in layer_sizes:
        for conv in conv_layers:
            NAME = f"{conv}-conv-{layer}-layer-{dense}-dense-{int(time.time())}"
            tensorboard = TensorBoard(log_dir=f'logs/{NAME}')

            model = Sequential()
                
            model.add(Conv2D(layer, (3,3), input_shape=X.shape[1:]))
            model.add(Activation("relu"))
            model.add(MaxPooling2D(pool_size=(2,2)))

            for i in range(conv-1):
                model.add(Conv2D(layer, (3,3)))
                model.add(Activation("relu"))
                model.add(MaxPooling2D(pool_size=(2,2)))
                
            model.add(Flatten())

            for j in range(dense):
                model.add(Dense(layer))
                model.add(Activation('relu'))

            model.add(Dense(1))
            model.add(Activation('sigmoid'))
            

            model.compile(
                loss='binary_crossentropy', 
                optimizer='adam', 
                metrics=['accuracy']
            )

            model.fit(X, y, batch_size=32, validation_split=0.3, epochs=10, callbacks=[tensorboard])
            
model.save('64x3-cnn.model')
```
